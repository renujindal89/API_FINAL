```python
Why API_KEY is required ❓
1️⃣ Identifies who is using the API

The API key tells TMDb:

Who you are
Which application is making the request
Without it, TMDb doesn’t know whether the request is authorized.
```

```python
Step-by-step flow
You send request with API key
TMDB server receives it
Server checks:
✅ Is this API key valid?
✅ Is it registered?
If valid → returns data
If invalid → error ❌
```

```python
Step-by-step process
1️⃣ You create an account

Example: on The Movie Database (TMDB)

2️⃣ You request an API key
Go to settings → API section
Click Generate API Key
3️⃣ Server generates the key

👉 Behind the scenes, the server creates something like:

842e8abdab255b000ba1aa62379231e5

4️⃣ Key is linked to you
In database:

API_KEY → User (Renu)

5️⃣ You use it in requests
params = {
    "api_key": "your_key_here"
}
```

```python
API Key = Your name at reception
Token = Temporary visitor pass with time limit
```

```python
import requests

API_KEY = "842e8abdab255b000ba1aa62379231e5"

url = "https://api.themoviedb.org/3/movie/popular"
params = {
    "api_key": API_KEY,
    "language": "en-US",
    "page": 1
}
```

```python
response = requests.get(url, params=params)
data = response.json()

print(data["results"][0]["title"])
```

```python
data["results"][0]
```

```python
import pandas as pd

df = pd.DataFrame(data["results"])
df[["title", "release_date", "vote_average"]].head()
```

```python
url = "https://api.themoviedb.org/3/movie/top_rated"

params = {
    "api_key": "842e8abdab255b000ba1aa62379231e5",
    "language": "en-US",
    "page": 1,
    "region": "IN"
}
```

```python
movies = data["results"]
```

```python

high_rated = [m for m in movies if m["vote_average"] > 8]
```

```python
Token (Access Token)

👉 A temporary permission key with more security.
More secure than API key
```

```python
Feature	                       API Key 🔑	       Token 🎟️
Purpose	                     Identify app	      Identify + authorize user
Security	                Less secure             	More secure
Expiry	                      Usually no expiry	      Expires (time-based)
Usage	                     Simple APIs	          Secure APIs (login required)
Sent in	                    Params / headers	          Mostly headers
```
